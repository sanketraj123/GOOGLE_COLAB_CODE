```python
basic = float(input("enter the basic salary: "))
hra = (basic*10)/100
Ta = (basic*5)/100

GS = basic+hra+Ta
print("gross salary is: ",GS)
tax=(GS*10)/100
NS=GS-tax
print("net salary is: ",NS)
```
